## How to label customers whose aggregated `amount_spent` is in a given quantile range?


#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

```python
# Aggregate the amount_spent for unique customers.
orders_percustomer_amount = orders.groupby('CustomerID').sum('amount_spent')
```

```python
orders_percustomer_amount
```

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

```python
orders_percustomer_amount[orders_percustomer_amount['amount_spent'] > orders_percustomer_amount['amount_spent'].quantile(0.95)]
```

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

We obtain vip_ids list and preferred_ids list.

```python
vip_ids = orders_percustomer_amount[orders_percustomer_amount['amount_spent'] > orders_percustomer_amount['amount_spent'].quantile(0.95)].index
```

```python
vip_ids
```

```python
preferred_ids = orders_percustomer_amount[(orders_percustomer_amount['amount_spent'] > orders_percustomer_amount['amount_spent'].quantile(0.75)) & (orders_percustomer_amount['amount_spent'] <= orders_percustomer_amount['amount_spent'].quantile(0.95))].index
```

```python
preferred_ids
```

We label in new columns.

```python
def label_vips(row):
    if row['CustomerID'] in vip_ids.values:

        return 1
    else:
        return 0
```

```python
orders['VIP'] = orders.apply(label_vips, axis=1)
```

```python
def label_preferred(row):

    if row['CustomerID'] in preferred_ids.values:
       
        return 1
    else:
        return 0
```

```python
orders['PreferredCustomers'] = orders.apply(label_preferred, axis=1)
```

We check if the size of the labeled ids matches the count of ids for each type.

```python
orders_customer_vips = orders[orders['VIP'] == 1].groupby('CustomerID')
orders_customer_vips = pd.DataFrame(orders_customer_vips)
orders_customer_vips.shape
```

```python
orders_customer_preferred = orders[orders['PreferredCustomers'] == 1].groupby('CustomerID')
orders_customer_preferred = pd.DataFrame(orders_customer_preferred)
orders_customer_preferred.shape
```

## Q2: How to identify which country has the most VIP Customers?

```python
# We group by customerid and country.
orders_percustomer_country_amount = orders.groupby(['CustomerID', 'Country']).sum('amount_spent').reset_index()
orders_percustomer_country_amount
```

There are ids that exist in more than one country:

```python
orders_percustomer_country_amount['CustomerID'].value_counts()
```

For example:

```python
orders_percustomer_country_amount[orders_percustomer_country_amount['CustomerID'] == 12417]
```

```python
# We filter by VIP_ids.
orders_perVIPcustomer_country_amount = orders_percustomer_country_amount[orders_percustomer_country_amount['CustomerID'].isin(vip_ids)]
```

```python
orders_perVIPcustomer_country_amount['Country'].value_counts()
```

**United Kingdom** has the largest sum of VIP customers.

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# We filter by VIP or Preferred ids.
orders_prefandVIP_country_amount = orders_percustomer_country_amount[(orders_percustomer_country_amount['CustomerID'].isin(vip_ids)) | (orders_percustomer_country_amount['CustomerID'].isin(preferred_ids))]
```

```python
orders_prefandVIP_country_amount['Country'].value_counts()
```

**United Kingdom** has the largest sum of combined VIP and Preferred customers.
